# Parser

```python
import bs4
from bs4 import BeautifulSoup
bs4.builder.builder_registry.builders
import os
from urllib.parse import urlparse
directory = '/Users/claudia_o_r/Python/Projects/malawi_news/urls_2019/' # files where our urls are
```

```python
i = 1
for filename in os.listdir(directory):
    print(filename)
    f = os.path.join(directory, filename)
    with open(f, 'r') as this_page:
        contents = this_page.read()
        soup = BeautifulSoup(contents, 'html')
        for tag in soup.find_all("meta"):
            if tag.get("property", None) == "og:title":
                title = tag.get("content", None)
            elif tag.get("property", None) == "article:published_time":
                published_time = tag.get("content", None)
            elif tag.get("property", None) == "og:url":
                url = tag.get("content", None)
                news_media = urlparse(url).netloc
                news_media = '.'.join(news_media.split('.')[-2:])
                news_media = news_media.replace(".com","")
        new_f = open("/Users/claudia_o_r/Python/Projects/malawi_news/headlines_dates_2019/"+news_media+"_"+str(i)+".txt", "x")# write text to file (in this case I am writing the urls and dates
        new_f.write(title)
        new_f.write(".")
        new_f.write(' ')
        new_f.write('\n')
        #since every newspaper has a different name for their div where the main content is, we need to define one for each
        print(published_time)
        new_f.write(published_time) #the following block writes the content, but I commented it out so to write only the title and date
        # for div in soup.find_all("div", class_="entry-content clearfix"):
        #     for p in div.find_all("p"):
        #         new_f.write(p.text)
        # for div in soup.find_all("div", class_="nyasa-content"):
        #     for p in div.find_all("p"):
        #         new_f.write(p.text)
        # for div in soup.find_all("div", class_="content-inner"):
        #     for p in div.find_all("p"):
        #         new_f.write(p.text)
        # for div in soup.find_all("div", class_="entry-content entry clearfix"):
        #     for p in div.find_all("p"):
        #         new_f.write(p.text)
        new_f.close()
    i+=1
    print("===================================")
    print("===================================")
```
